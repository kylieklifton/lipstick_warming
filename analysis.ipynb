# Lipstick Warming Analysis

Analyzing MAC lipstick undertone shifts over time.

```python
import pandas as pd
```

## Load MAC Shade Data

Data scraped from Temptalia via Wayback Machine (Nov 2024 archive).

```python
shades = pd.read_csv('data/mac_shades.csv')
shades.head(10)
```

## Undertone Distribution

What percentage of MAC's lineup is warm vs cool?

```python
undertone_counts = shades['undertone'].value_counts()
undertone_counts
```

```python
# Calculate percentages
total = len(shades[shades['undertone'].isin(['warm', 'cool', 'neutral'])])
warm_pct = (undertone_counts.get('warm', 0) / total * 100).round(1)
cool_pct = (undertone_counts.get('cool', 0) / total * 100).round(1)
neutral_pct = (undertone_counts.get('neutral', 0) / total * 100).round(1)

f"Warm: {warm_pct}% | Cool: {cool_pct}% | Neutral: {neutral_pct}%"
```

## Reformulation Evidence

Documented cases where MAC shades shifted warmer over time.

```python
reformulations = pd.read_csv('data/reformulation_evidence.csv')
reformulations[['shade_name', 'original_undertone', 'current_undertone', 'mac_response']]
```

## Temperature Change Data

Quantified undertone shift for reformulated shades.

```python
temp_change = pd.read_csv('data/temperature_change_mock.csv')
temp_change
```

```python
# Average temperature increase
avg_change = temp_change['change'].mean()
f"Average warming shift: +{avg_change:.1f} points on 5-point scale"
```

## Key Finding

MAC's lipstick lineup skews warm (56% warm vs 33% cool), and several classic shades have documented shifts toward warmer undertones. MAC has acknowledged this by releasing "Cool" versions of popular shades like Spice, Velvet Teddy, and others.
